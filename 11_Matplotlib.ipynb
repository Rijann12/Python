#Lables an titles

```python
x=np.array([80,85,90,95,100,105,110,120])
y= np.array([240,250,260,270,250,280,300,210])
plt.figure(figsize=(3,3))
plt.plot(x,y)

plt.xlabel("Average Pulse", fontsize=8)
plt.ylabel("Calorie Burnage", fontsize=8)
plt.title("sports watch data", color='red')
plt.show()
```

#Gridlines

```python
x=np.array([80,85,90,95,100,105,110,120])
y= np.array([240,250,260,270,250,280,300,210])
plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.grid(axis='y',linewidth=2,color='red')
plt.title("sports watch data", color='red')
plt.xlabel("Average Pulse", fontsize=8)
plt.ylabel("Calorie Burnage", fontsize=8)
plt.show()
```

#Scatter plot

```python
x1= np.array([250,26,23,23,29,50])
y1= np.array([1,2,9,8,20,50])

plt.figure(figsize=(3,3))
plt.scatter(x1,y1,label='Exp 1')

x2= np.array([5,26,2,23,2,50])
y2= np.array([55,22,98,8,20,50])

plt.scatter(x2,y2,label='Exp 2')
plt.legend()

plt.show()
```

```python
# importing package
import matplotlib.pyplot as plt

# create data
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]

# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.show()
```

```python
y= np.array([5,10,15,20])
labels=["apple","banana","peach","orange"]
explode=(0.1,0,0,0)
plt.figure(figsize=(6,6))
plt.pie(y,labels=labels,explode=explode)
plt.show()
```
