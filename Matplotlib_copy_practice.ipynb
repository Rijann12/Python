#LABEL AN TITLES


```python
x=np.array([22,66,88,89,32,100,67,12])
y=np.array([423,455,23,45,222,123,186,232])
plt.figure(figsize=(3,3))
plt.plot(x,y)

plt.xlabel("Average Pulse", fontsize=8)
plt.ylabel("Calorie Burn", fontsize=8)
plt.title("sports watch data", color='red')
plt.show()
```

#GRIDLINES


```python
x=np.array([22,66,88,89,32,100,67,12])
y=np.array([423,455,23,45,222,123,186,232])
plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.grid(axis='y',linewidth=2,color='red')
plt.xlabel("Average Pulse", fontsize=8)
plt.ylabel("Calorie Burn", fontsize=8)
plt.title("sports watch data", color='red')
plt.show()
```

```python
#importing package
import matplotlib.pyplot as plt

#create data
x=['a','b','c','d']
y1=[10,20,30,40]
y2=[23,12,34,55]
#plot bars in stack manner
plt.bar(x,y1,color='r')
plt.bar(x,y2,bottom=y1,color='b')
plt.show()
```

```python
y=np.array([5,10,12,44])
labels=["apple","banana","orange","peach"]
explode=(0.1,0,0,0)
plt.figure(figsize=(5,5))
plt.pie(y,labels=labels,explode=explode)
plt.show()
```
